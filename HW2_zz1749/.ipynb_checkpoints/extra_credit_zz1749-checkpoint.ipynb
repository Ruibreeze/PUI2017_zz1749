```python
data = pd.read_csv(os.getenv("DFDATA")+"/724b-yxwa/1476480727/724b-yxwa.csv")
```

```python
data.head()
```

```python
data = pd.DataFrame({"Created Date" : list(data.iloc[:,1]),"Incident Zip":list(data.iloc[:,8])})
```

```python
data.head()
```

```python
new_x = []
for i in range(len(list(data["Created Date"]))):
    new_x.append(datetime.datetime.strptime(list(data["Created Date"])[i],'%m/%d/%Y %I:%M:%S %p'))
print(new_x)
```

```python
data["Created Date"] = new_x
```

```python
data.head()
```

```python
data.plot(data["Created Date"],["Incident Zip"],rot=90,ylim=[11200,11250])

plt.xlabel('Created Date')
plt.ylabel('Incident Zip')
plt.title('Reducted Dataframe')
plt.legend(loc="upper right")
plt.show()
```
